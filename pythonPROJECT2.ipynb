```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
%matplotlib inline
sns.set(color_codes=True)
import matplotlib.pyplot as plt
```

```python
file_path = '/content/Test.csv'
df = pd.read_csv(file_path, encoding='utf-8')
```

```python
# assuming df is your DataFrame
sns.barplot(x=df['humidity'], y=df['temperature'])
plt.show(block=True)  # Use plt.show() instead of matplotlib.show()
```

```python
sns.displot(df['humidity'])
```

```python
#assuming df is your dataframe
sns.jointplot(x='humidity', y='temperature',data=df,color='r')
```

```python
#assuming df is your dataframe
sns.jointplot(x='humidity', y='temperature',data=df, kind="hex", color='g')
```

```python
#assuming df is your dataframe
sns.jointplot(x='humidity', y='temperature',data=df, kind="kde",color='r')
```

```python
#assuming df is your dataframe
sns.pairplot(df[['humidity','temperature','air_pollution_index']])
```

```python
#assuming df is your dataframe
sns.stripplot(x='weather_type', y='temperature',data=df, color='g')
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='humidity', y='temperature',data=df,hue=df['weather_type'])
plt.show()
```

```python
#assuming df is your dataframe
sns.swarmplot(x='humidity', y='temperature',data=df, color='g')
```

```python
sns.countplot(df['weather_type'], color='y')
```

```python
#assuming df is your dataframe
sns.pointplot(x='humidity', y='temperature',data=df, color='r', hue=df['weather_type'])
```
